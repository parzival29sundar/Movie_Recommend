```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
#genre
#id
#keywords
#title
#overview
#popularity
#cast
#crew

movies= movies[['id' ,  'original_title', 'overview','genres', 'keywords', 'cast','crew']]
```

```python
movies.head()
```

```python
#We are removing all the values which contain NULL.
```

```python
movies.dropna(inplace =True)
```

```python
movies.isnull().sum()
```

```python
#Checking duplicate data
```

```python
movies.duplicated().sum()
```

```python
import ast
ast.literal_eval
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#we only taking 4 cast name from the top, they are more relavent for recommmandation.

def convert4(obj):
    L=[]
    counter =0
    for i in ast.literal_eval(obj):
        if counter !=4:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L    
```

```python
movies['cast'] = movies['cast'].apply(convert4)
```

```python
movies.head()
```

```python
# for crew , we are taking job(in crew dictionary) whose value is director, for recommandation.
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L         
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#now we are converting the overview column in string 
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#now we have to remove spaces, from the name , as there could be more than 1 sam , so we need to remove spaces and create SAMWORTHINGTON

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x]) #replace " " space with "" nothing
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
#we are concatinating alll the 4 columns and creating a single tags column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
#as the tags column is created , we can delete the other 4 columns
new_df = movies[['id' ,  'original_title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
#converting into lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=4000, stop_words='english')  
```

```python
#sklearn countvectorization will return a sparse matrix , we need to convert it into a numpy array
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['original_title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].original_title)
        
    
```

```python
recommend('Batman Begins')
```
